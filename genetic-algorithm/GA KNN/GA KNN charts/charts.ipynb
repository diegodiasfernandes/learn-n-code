```python
import matplotlib.pyplot as plt

def plot_graph(y_values):
    if len(y_values) != 10:
        raise ValueError("The input array must contain exactly 10 values.")

    x_values = list(range(1, 11))

    plt.figure(figsize=(12, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

    plt.title('GA Accuracies Through Generations')
    plt.xlabel('Generations')
    plt.ylabel('Accuracy')

    plt.xticks(x_values)
    plt.grid(False)
    plt.show()

# Example usage
y_values = [0.724074074074074, 0.7722222222222223, 0.7962962962962962,
            0.8055555555555556, 0.7962962962962962, 0.8277777777777778,
            0.8296296296296296, 0.837037037037037, 0.8240740740740741, 0.837037037037037]
plot_graph(y_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_graph_with_regression(y_values):
    if len(y_values) != 10:
        raise ValueError("The input array must contain exactly 10 values.")

    x_values = np.arange(1, 11)

    # Calculate the linear regression
    coefficients = np.polyfit(x_values, y_values, 1)  # 1 indicates a linear fit (degree 1)
    polynomial = np.poly1d(coefficients)
    regression_y_values = polynomial(x_values)

    # Plot the original data points
    plt.figure(figsize=(12, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Generation Accuracy Mean')

    # Plot the regression line
    plt.plot(x_values, regression_y_values, linewidth=3, linestyle='--', color='r', label='Accuracy Regression')

    # Add titles and labels
    plt.title('GA Accuracy Tendency Through Generations')
    plt.xlabel('Generations')
    plt.ylabel('Accuracy')

    plt.xticks(x_values)
    plt.grid(False)
    plt.legend()
    plt.show()

# Example usage
y_values = [0.724074074074074, 0.7722222222222223, 0.7962962962962962,
            0.8055555555555556, 0.7962962962962962, 0.8277777777777778,
            0.8296296296296296, 0.837037037037037, 0.8240740740740741, 0.837037037037037]

plot_graph_with_regression(y_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def find_best_polynomial_degree(x, y, max_degree):
    best_degree = 1
    best_error = float('inf')

    for degree in range(1, max_degree + 1):
        coefficients = np.polyfit(x, y, degree)
        polynomial = np.poly1d(coefficients)
        y_pred = polynomial(x)
        error = mean_squared_error(y, y_pred)

        if error < best_error:
            best_error = error
            best_degree = degree

    return best_degree

def plot_graph_with_best_fit(y_values, max_degree=5):
    if len(y_values) != 10:
        raise ValueError("The input array must contain exactly 10 values.")

    x_values = np.arange(1, 11)

    # Find the best polynomial degree
    best_degree = find_best_polynomial_degree(x_values, y_values, max_degree)

    # Calculate the best polynomial regression
    coefficients = np.polyfit(x_values, y_values, best_degree)
    polynomial = np.poly1d(coefficients)
    regression_y_values = polynomial(x_values)

    # Plot the original data points
    plt.figure(figsize=(12, 6))
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Generation Accuracy Mean')

    # Plot the best fit regression line
    plt.plot(x_values, regression_y_values, linewidth=3, linestyle='--', color='r', label='Accuracy Regression')

    # Add titles and labels
    plt.title('GA Accuracy Tendency Through Generations')
    plt.xlabel('Generations')
    plt.ylabel('Accuracy')

    plt.xticks(x_values)
    plt.grid(False)
    plt.legend()
    plt.show()

# Example usage
y_values = [0.724074074074074, 0.7722222222222223, 0.7962962962962962,
            0.8055555555555556, 0.7962962962962962, 0.8277777777777778,
            0.8296296296296296, 0.837037037037037, 0.8240740740740741, 0.837037037037037]
plot_graph_with_best_fit(y_values, max_degree=3)
```
